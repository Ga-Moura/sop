```python
import pandas as pd
```

```python
df1 = {

'objeto de locação':['0001001','0001002','0001006','0001007','0001008','0001009'] ,
 'retalhonamento' :[None,'0001004','0001004','0001003',None,'0001001']

}

df1 = pd.DataFrame(df1)
```

```python
df2 = {

'objeto de locação':['0001001','0001003','0001004','0001007'] ,
 'informação' :['G1','G3','G4','GTESTES'],
 'informação2':['G5','G6','G7','GTESTES'],
 'informação3' :['G8','G9','G10','G11']

}

df2 = pd.DataFrame(df2)
```

```python
def merge_multiplo(df1, df2, key_left1, key_left2, key_right, columns_list):
    
    #Criar um primeiro DF inicial que não se refira a base principal
    df_ini = df1.merge(df2[columns_list], left_on=key_left1, right_on=key_right, how='left')
    
    
    #Criação de um DF intermediário -> Esse DF terá dados com colunas confusas e cheias de informação, iremos reorganizá-la para o DF inicial
    df_inter = df1.merge(df2[columns_list],left_on=key_left2,right_on=key_right,how='left')


    #Definindo as linhas vazias da primeira coluna
        #Pensar em uma lógica para olhar para todas as colunas e ver se elas estão vazias#

    colunas_vazias = []

    for coluna in df_ini.columns:
        
        if df_ini[coluna].isna().any() and coluna != key_left2:
            colunas_vazias.append(coluna)
    
    print(f"Coluans identificadas com valores que podem ser preenchidos com a coluna {key_left2}: {colunas_vazias}")
    print('_'*150)

    for coluna_vazia in colunas_vazias:
        
        for coluna_inicial in df_ini.columns:
            if coluna_vazia == coluna_inicial:
                print(f'Colunas com nomes iguais são:{coluna_inicial}')

                df_ini[coluna_vazia] = df_ini[coluna_vazia].fillna(df_inter[coluna_vazia]) 


         
      
    return df_ini

    
  
```

```python
df_fim = merge_multiplo(df1,df2,'objeto de locação','retalhonamento','objeto de locação',['objeto de locação','informação','informação2','informação3'])

df_fim
```
